## 7. Reload the dataset so that null values are correctly interpreted (repeat section 4, delete columns).

```python
df_airports_2=pd.read_csv('airports.csv', sep=";",na_values='\\N');df_airports_2.head()
```

```python
df_airports_2.info()
```

```python
df_airpots_2_clean=df_airports_2.drop(['AirportID','Latitude','Longitude','Altitude'],axis=1).copy();df_airpots_2_clean.head()
```

```python
df_airpots_2_clean.info()
```

## 8. Checks the entire dataframe for null values.

```python
values_null=df_airpots_2_clean.isna().sum().sum(); values_null
```

```python
df_airpots_2_clean[df_airpots_2_clean.columns].isna().sum()
```

## 9. Overwrites the null values of the IATA and ICAO columns with the value 'UNKNOWN'

```python
df_airpots_2_clean.IATA=df_airpots_2_clean.IATA.fillna('UNKNOWN')
df_airpots_2_clean.IATA=df_airpots_2_clean.ICAO.fillna('UNKNOWN')
df_airpots_2_clean[["IATA","ICAO"]]
```

## 10. Changes the type of the DST and TZ variables to categorical.

```python
df_airpots_2_clean["DST"]=pd.Categorical(df_airpots_2_clean["DST"])
df_airpots_2_clean["TZ"]=pd.Categorical(df_airpots_2_clean["TZ"])
df_airpots_2_clean
```

```python
df_airpots_2_clean.info()
```
