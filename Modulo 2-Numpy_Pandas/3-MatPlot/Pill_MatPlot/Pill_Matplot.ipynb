```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
os.getcwd()
```

```python
df_sales=pd.read_csv("company_sales_data.csv");df_sales
```

## 1. Read Total profit of all months and show it using a line plot

```python
plt.plot(df_sales.total_profit)
plt.show()
```

## 2. Get total profit of all months and show line plot with the following Style properties:
    a)-Line Style dotted and Line-color should be red
    b)-Show legend at the lower right location.
    c)-X label name = Month Number
    d)-Y label name = Sold units number
    e)-Add a circle marker.
    f)-Line marker color as read
    g)-Line width should be 3


```python
plt.plot(df_sales.total_profit,label="total_amounts",ls="dotted",c="r",marker="o",lw=3)
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
plt.legend(loc="upper left")
plt.show()
```

## 6. Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
indice = range(0, len(df_sales))
fig, ax = plt.subplots(2,1, figsize=(10, 10))

ax[0].bar(indice,df_sales.facecream,color="red",label = "facecream")
ax[1].bar(indice,df_sales.facewash,color="blue",label = "facewash")

plt.show()
```
